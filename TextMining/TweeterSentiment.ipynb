```python
class TwitterSentiment(object):         # generic twitter class for sentiment Analysis
    def __init__(self):         # Class constructor
        
        # key and token from the twitter dev console
        consumer_key = '1wvXjBvgHNmX8BbSxzvL3QP3t'
        consumer_secret = 'zf44jbi23fH8ljrGi0QExKi0nnfSyeIxkdiOD2mM5VObbvp1xA'
        access_token = '59418573-f96OjTLMTOgPnYdfFUO5C9jibWob3qeLeANQx6C5z'
        access_token_secrate = 'sYUa7xxe1nQiBhnVaLSqKUj6cxYFFLuIekIkeSY0Ctb6a'
        # Attempt authentication
        try:
            # Create OAuthhandler Object
            self.auth = OAuthHandler(consumer_key, consumer_secret)
            
            # Set Access Token and secrate
            self.auth.set_access_token(access_token,access_token_secrate)
            
            #create tweepy API object to fetch tweets
            self.api = tweepy.API(self.auth)
        except:
            print("Tweeter API Authentication failure.")
            
    def clean_tweets(self, tweet):
        '''Utility function to clean tweet text by removing links, special characters, using simple regx statement'''
        return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
    def get_tweet_sentiment(self, tweet):
        '''Utility function to classify the sentiment of passed tweet using TextBlob sentiment method'''
        # Create a TextBlob object of passed tweet text
        analysis = TextBlob(self.clean_tweets(tweet))
        # Set sentiment
        if analysis.sentiment.polarity > 0:
            return 'positive'
        if analysis.sentiment.polarity == 0:
            return 'neutral'
        if analysis.sentiment.polarity < 0:
            return 'negative'
    def get_tweets(self, query, count=20):
        '''main function to fetch the tweets from the tweeter and parsed them'''
        # Empty tweets to the parsed tweets
        tweets = []
        
        try:
            # Call tweeter API to fetch tweets
            fetched_tweets = self.api.search(q = query, count = count)
            
            #parsing tweets one-by-one
            for tweet in fetched_tweets:
                #Empty dictionary to store required parameter of a tweet
                parsed_tweet = {}
                
                # Saving text of tweets
                parsed_tweet['text'] = tweet.text
                
                # Saving sentiment of a tweet
                parsed_tweet['sentiment'] = self.get_tweet_sentiment(tweet.text)
                
                # Appending parsed tweet to tweets list
                if tweet.retweet_count > 0:
                    
                    # if tweets has retweets, ensure that it is appended only once
                    if parsed_tweet not in tweets:
                        tweets.append(parsed_tweet)
                else:
                    tweets.append(parsed_tweet)
            # return parsed tweets        
            return tweets
        except :
            print("Error :")
def main():
    #Creating object of a TwitterSentiment class
    api = TwitterSentiment()
    
    # calling function to get tweets
    tweets = api.get_tweets(query = '#IBMAnalytics', count = 200)
    
    # picking positive tweets from tweets
    ptweets = [tweet for tweet in tweets if tweet['sentiment'] == 'positive']
    
            
    # percentage of positive tweets 
    print("Percentage of positive tweets :", ((len(ptweets) / len(tweets))*100))
    
    # Picking negative tweets
    ntweets = [tweet for tweet in tweets if tweet['sentiment'] == 'negative']
    
    # Percentage of negative tweets
    print("Percentage of Negative tweets :", (len(ntweets) / len(tweets))* 100)
    
    # printing first 5 positive tweets
    print("\n\n positive tweets :")
    for tweet in ptweets[:]:
        print(tweet['text'])
        
    # print first 5 negative tweets
    print("\n\n Negative tweets : ")
    for tweet in ntweets[:]:
        print(tweet['text'])
    
if __name__ == '__main__':
    # calling main function
    main()
        
```
